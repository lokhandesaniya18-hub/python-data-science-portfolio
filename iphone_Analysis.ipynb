```python
import pandas as pd
```

```python
df=pd.read_csv(r"iphone_sales.csv")
df
```

```python
df['Price']=pd.to_numeric(df['Price'])
df['Quantity']=pd.to_numeric(df['Quantity'])
df
```

```python
df['TotalAmount']=df['Price']*df['Quantity']
df['TotalAmount']
```

```python
df
```

```python

total_revenue=df['TotalAmount'].sum()
print(f"Total iphone Revenue: R{total_revenue}")
```

```python
model_sales=df.groupby('Model')['Quantity'].sum().sort_values(ascending=False)
print("Best Selling iphone Models:")
model_sales
```

```python
City_revenue=df.groupby('City')['TotalAmount'].sum().sort_values(ascending=False)
print("city-wise Revenue")
City_revenue
```

```python
premium_phones=df[df['Price']>100000]
print("premium phones > 1 Lakh:")
premium_phones[['Model','Price','City']]
```

```python
df.to_csv('iphone_sales_report.csv',index=False)
print("PROJECT COMPLETE! 'iphone_sales_report.csv'")
```
